# 미국에서 연재하는 한국 작품 데이터프레임 생성

```python
import pandas as pd
```

```python
# 1. korea 데이터 불러오고 결측치 제거
korea = pd.read_csv("data/korea_add.csv")
```

```python
korea
```

```python
# 필요없는 칼럼 제거
korea.drop(["Unnamed: 0","Unnamed: 11"],inplace=True,axis=1)
```

```python
korea.info()
```

```python
# 결측치 제거
korea.isnull().sum()
korea.dropna(inplace=True)
```

```python
korea.reset_index(drop=True,inplace=True)
```

```python
korea.info()
```

```python
korea.tail()
```

```python
# 2. america 데이터 불러오고, data frame 생성
```

```python
america = pd.read_csv("data/america.csv")
```

```python
## 한국원작의 미국 제목이 들어간 america_title list 만들기
korea_dict = korea.to_dict()
korea_val = korea_dict["america_title"].values()
data_title_list = list(korea_val)
```

```python
data_title_list
```

```python
## 미국 제목 리스트 만들기
america_dict = america.to_dict()
america_val = america_dict["title"].values()
america_title_list = list(america_val)
```

```python
len(america_title_list)
```

```python
ranks = []
for title in data_title_list:
    title = title.replace("\xa0","") #전처리
    if title in america_title_list:
        rank = america[america["title"] == title]["rank"].values[0]
        ranks.append(rank)
    else:
        ranks.append(0)
```

```python
len(ranks)
```

```python
len([1 for rank in ranks if rank ==0])
```

```python
## 데이터 프레임 생성

data = korea.copy()
```

```python
data.info()
```

```python
data["america_rank"] = ranks
```

```python
data[data["america_rank"] == 0]
```

```python
add_ranks = [6, 20, 60, 0, 41, 2, 70, 0, 0, 7, 0] # 0은 미국 네이버웹툰에 연재하지만, 시작한지 얼마 되지 않아 순위 집계X
```

```python
data.loc[data["america_rank"]==0,"america_rank"] = add_ranks
```

```python
## rank가 0인 행 제거
data.drop([47,79,80,88], axis=0,inplace=True)
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
## 미국에 연재하는 작품들이므로 isamerica는 필요 없는 칼럼, 삭제
data.drop(["isamerica"],inplace=True,axis=1)
```

```python
data.info()
```

```python
# 3. 저장

data.to_excel('data.xlsx')
data.to_csv('data.csv')
```
